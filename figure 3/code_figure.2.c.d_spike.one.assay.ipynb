```python
import pandas as pd
import numpy as np
import seaborn as sbn
import matplotlib.pyplot as plt

from datetime import datetime
current_date = datetime.now()
formatted_date = current_date.strftime("%Y.%m.%d")
```

```python
root = '/Users/grantongo/Desktop/nature methods/figures'
root_fig = '/Users/grantongo/Desktop/nature methods/figures/figure 3'

df_sm = pd.read_csv(f'{root_fig}/data_spike_1_in_10ngmL_{formatted_date}.csv') #, header=None)
df_mm = pd.read_csv(f'{root_fig}/data_leave_some_out_100pgmL{formatted_date}.csv') #, header=None)

df_sm = df_sm.set_index('Protein')
df_mm = df_mm.set_index('Protein')

cmap= sbn.blend_palette(['white','#ffdf22','#76C8A5','royalblue','#5E4FA2'], n_colors=400)
```

```python
# spike 1 in (1 ng/mL)

fig, axs = plt.subplots(1, 1, figsize=(20,14))
axs = sbn.heatmap(df_sm.T,
                  xticklabels=False, 
                  yticklabels=False,
                  square=True,
                  cmap = cmap,
                  vmin=-5,vmax=120,
                  rasterized=True
                 )
axs.set_xlabel('Spiked Antigen',fontsize=20)
axs.set_ylabel('nELISA Sensor',fontsize=20)
fig.savefig(f'{root_fig}/svg_figure.3.c_spike1in_{formatted_date}.svg')
fig.savefig(f'{root_fig}/png_figure.3.c_spike1in_{formatted_date}.png',dpi=200)
```

```python
# leave some out (100 pg/mL per target)

fig, axs = plt.subplots(1, 1, figsize=(7,14))
axs = sbn.heatmap(df_mm.T,
                  xticklabels=False, 
                  yticklabels=False,
                  cmap= cmap,
                  vmin=-5,vmax=120,
                  rasterized=True
                 )
axs.set_ylabel('nELISA Sensor',fontsize=20)
axs.set_xlabel('Antigen Pools',fontsize=20)
fig.savefig(f'{root_fig}/svg_figure.3.d_leavesomeout_{formatted_date}.svg')
fig.savefig(f'{root_fig}/png_figure.3.d_leavesomeout_{formatted_date}.png',dpi=200)
```
